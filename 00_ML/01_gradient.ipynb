```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. 
x = np.arange(0, 4, 0.01)  # x0~40.013D
y = np.arange(0, 4, 0.01)
X, Y = np.meshgrid(x, y)  # 2D

# 2.  z  (dx, dy)
z = X**2 + X*Y + Y**2  # 
dx = 2*X + Y           # xx
dy = X + 2*Y           # yy

# 3. 23D + 2D+
fig = plt.figure(figsize=(14, 6))

# 13D
ax1 = fig.add_subplot(121, projection='3d')
surface = ax1.plot_surface(
    X, Y, z,
    cmap='viridis',  # z
    alpha=0.8,       # 
    edgecolor='none',# 
    antialiased=True # 
)
# 3D
ax1.set_xlabel('X Axis', fontsize=12)
ax1.set_ylabel('Y Axis', fontsize=12)
ax1.set_zlabel('Z Axis', fontsize=12)
ax1.set_title('3D Surface: $z = x^2 + xy + y^2$', fontsize=14, pad=20)
fig.colorbar(surface, ax=ax1, shrink=0.6)  # z

# 22D + 
ax2 = fig.add_subplot(122)
# z
contour = ax2.contour(
    X, Y, z,
    levels=20,        # 
    cmap='viridis',
    linewidths=1.2
)
ax2.clabel(contour, inline=True, fontsize=8)  # 

# 401
step = 40
X_sparse = X[::step, ::step]
Y_sparse = Y[::step, ::step]
dx_sparse = dx[::step, ::step]
dy_sparse = dy[::step, ::step]

# quiver
ax2.quiver(
    X_sparse, Y_sparse,  # 
    dx_sparse, dy_sparse, # x/y
    color='red',         # 
    angles='xy',         # x/y
    scale_units='xy',    # x/y
    scale=20,            # 
    headwidth=6,         # 
    alpha=0.7            # 
)
# 2D
ax2.set_xlabel('X Axis', fontsize=12)
ax2.set_ylabel('Y Axis', fontsize=12)
ax2.set_title('Contour + Gradient Vector Field', fontsize=14)
ax2.grid(True, alpha=0.3)
ax2.set_aspect('equal')  # 

plt.tight_layout()
plt.show()
```
